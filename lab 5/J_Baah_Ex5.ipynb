```python
# Import the needed libraries
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import BorderlineSMOTE,SMOTE,ADASYN 


from numpy import where
import seaborn as sns
from collections import Counter
```

```python
# Visualization of the dataset before oversampling techniques are used
sns.scatterplot(data=df,x=df["feature_x"],y=df["feature_y"],hue=df['class'])
```

```python
# Using another visualization to depict the imbalance
sns.countplot(df['class'])
```

```python
smote = SMOTE()
new_x,new_y = smote.fit_resample(df[["feature_x","feature_y"]],df["class"].astype('int'))
```

```python
# Put results back in dataframe to plot on a scatter plot
oversampled_df = pd.DataFrame(new_x,columns=["feature_x","feature_y"])
oversampled_df["class"] = new_y
sns.scatterplot(data=oversampled_df,x=oversampled_df["feature_x"],y=oversampled_df["feature_y"],hue=oversampled_df["class"])
```

```python
# Using another visualization to depict the change
sns.countplot(oversampled_df['class'])
```

```python
# Using the borderline SMOTE technique

borderline_smote = BorderlineSMOTE()
border_x,border_y = borderline_smote.fit_resample(df[["feature_x","feature_y"]],df["class"].astype('int'))
```

```python
# Put results back in dataframe to plot on a scatter plot
oversampled_border_df = pd.DataFrame(border_x,columns=["feature_x","feature_y"])
oversampled_border_df["class"] = border_y
sns.scatterplot(data=oversampled_border_df,x=oversampled_border_df["feature_x"],y=oversampled_border_df["feature_y"],hue=oversampled_border_df["class"])
```

```python
# Using another visualization to depict the change
sns.countplot(oversampled_border_df['class'])
```

```python
# Using the ADASYN technique

adasyn = ADASYN()
adasyn_x,adasyn_y = adasyn.fit_resample(df[["feature_x","feature_y"]],df["class"].astype('int'))
```

```python
# Put results back in dataframe to plot on a scatter plot
oversampled_adasyn_df = pd.DataFrame(adasyn_x,columns=["feature_x","feature_y"])
oversampled_adasyn_df["class"] = adasyn_y
sns.scatterplot(data=oversampled_adasyn_df,x=oversampled_adasyn_df["feature_x"],y=oversampled_adasyn_df["feature_y"],hue=oversampled_adasyn_df["class"])
```

```python
# Using another visualization to depict the change
sns.countplot(oversampled_adasyn_df['class'])
```
